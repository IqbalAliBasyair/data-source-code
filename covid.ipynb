```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
nama_file = 'Data Vaksin Covid 19.csv'
data = pd.read_csv(nama_file) 
data
```

```python
is_na = data.isna().any(axis=1)
data.loc[is_na] 
```

```python
duplicated = data.append(data.iloc[[-2]])
duplicated
```

```python
duplicated_data = duplicated.duplicated()
duplicated_data
```

```python
new = duplicated.drop_duplicates()
new
```

```python
N = 7
vaksin1 = new['Persentase Vaksin 1'][0:7]
vaksin2 = new['Persentase Vaksin 2'][0:7]
provinsi = new['Provinsi'][0:7]
ind = np.arange(N) # the x locations for the groups

width = 0.35 # the width of the bars
fig, ax = plt.subplots()
vaksins1 = ax.bar(ind, vaksin1, width, color='#ffad00')
vaksins2 = ax.bar(ind+width, vaksin2, width, color='#9b3c38')

# add some text for labels, title and axes ticks
ax.set_xticks(ind+width)
ax.set_xticklabels( (provinsi))
ax.legend( (vaksins1[0], vaksins2[0]), ('Vaksinasi Covid Dosis 1', 'Vaksinasi Covid Dosis 2') )

def autolabel(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        hcap = "  "+str(height)+"%"
        ax.text(rect.get_x()+rect.get_width()/2., height, hcap,ha='center', va='bottom', rotation="vertical")

autolabel(vaksins1)
autolabel(vaksins2)

plt.rcParams["figure.figsize"] = (20,5)
plt.show()
```
